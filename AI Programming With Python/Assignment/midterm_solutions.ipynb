```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

## Starter Code (Run First)

```python
np.random.seed(24)
num_employees = 50

emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

data.to_csv("staff_data.csv", index=False)
print("Dataset created and saved as staff_data.csv")
print(data.head())
```

---
# Section 1: NumPy — 5 Questions

## Q1

```python
# Q1 — NumPy arrays from emp_ages and emp_salary columns

ages_array   = data["emp_age"].to_numpy()
salary_array = data["emp_salary"].to_numpy()

print("=== emp_ages array ===")
print(ages_array)
print(f"  Data type       : {ages_array.dtype}")
print(f"  Dimensions      : {ages_array.ndim}")
print(f"  Shape           : {ages_array.shape}")
print(f"  Total elements  : {ages_array.size}")

print()
print("=== emp_salary array ===")
print(salary_array)
print(f"  Data type       : {salary_array.dtype}")
print(f"  Dimensions      : {salary_array.ndim}")
print(f"  Shape           : {salary_array.shape}")
print(f"  Total elements  : {salary_array.size}")
```

## Q2

```python
# Q2 — Max, min, average salary; average age

max_salary = np.max(salary_array)
min_salary = np.min(salary_array)
avg_salary = np.mean(salary_array)
avg_age    = np.mean(ages_array)

print(f"Maximum Salary   : {max_salary}")
print(f"Minimum Salary   : {min_salary}")
print(f"Average Salary   : {avg_salary:.2f}")
print(f"Average Age      : {avg_age:.2f}")
```

## Q3

```python
# Q3 — Filter: age between 25–40 AND salary > 50,000

condition = (ages_array >= 25) & (ages_array <= 40) & (salary_array > 50000)

filtered_ages    = ages_array[condition]
filtered_salaries = salary_array[condition]

print("Filtered Ages   :", filtered_ages)
print("Filtered Salaries:", filtered_salaries)
print(f"Number of employees satisfying the condition: {np.sum(condition)}")
```

## Q4

```python
# Q4 — Adjust ages without modifying original
# +4 if age < avg_age, +2 if age >= avg_age

average_age  = np.mean(ages_array)
updated_ages = np.where(ages_array < average_age,
                        ages_array + 4,
                        ages_array + 2)

print(f"Average age used for condition : {average_age:.2f}")
print("Original ages array :", ages_array)
print("Updated  ages array :", updated_ages)
```

## Q5

```python
# Q5 — Tech roles get 8% salary raise; calculate totals

tech_roles = [
    "Software Developer",
    "Backend Developer",
    ]
is_tech    = np.isin(emp_roles, tech_roles)

new_salaries   = np.where(is_tech, salary_array * 1.08, salary_array)

old_total      = np.sum(salary_array)
new_total      = np.sum(new_salaries)
total_increase = new_total - old_total

print(f"Old Total Salary Expense : {old_total:,.2f}")
print(f"New Total Salary Expense : {new_total:,.2f}")
print(f"Overall Increase         : {total_increase:,.2f}")
```

# Section 2: Pandas — 5 Questions

## Q6

```python
# Q6 — Load CSV, display first 3 rows, info, and describe

df = pd.read_csv("staff_data.csv")

print(" First 3 rows ")
print(df.head(3))

print("\n Dataset Info ")
df.info()

print("\n Summary Statistics")
print(df.describe())
```

## Q7

```python
# Q7 — Select emp_id, emp_age, role; filter age 25–40; sort by emp_id; show last 3

filtered_df = (
    df[["emp_id", "emp_age", "role"]]
    .query("emp_age >= 25 and emp_age <= 40")
    .sort_values("emp_id")
)

print("Last 3 rows of filtered & sorted result")
print(filtered_df.tail(3))
```

## Q8

```python
# Q8 — Filter employees in HR roles (HR Manager OR HR Executive)

hr_df = df[df["role"].str.contains("HR", case=False)]

print("HR Employees")
print(hr_df)
print(f"\nTotal number of HR employees: {len(hr_df)}")
```

## Q9

```python
# Q9 — Sort by emp_age ascending, then emp_salary descending; show top 5

sorted_df = df.sort_values(
    by=["emp_age", "emp_salary"],
    ascending=[True, False]
)

print("Top 5 employees after sorting")
print(sorted_df[["emp_id", "role", "emp_salary"]].head(5))
```

## Q10

```python
# Q10 — Apply salary raises using .loc; print role, previous_salary, new_salary, increase

df["previous_salary"] = df["emp_salary"].copy()


tech_mask  = df["role"].isin(["Software Developer", "Backend Developer"])
other_mask = ~tech_mask


df.loc[tech_mask,  "emp_salary"] = (df.loc[tech_mask,  "emp_salary"] * 1.12).round(2)
df.loc[other_mask, "emp_salary"] = (df.loc[other_mask, "emp_salary"] * 1.05).round(2)


df["new_salary"]        = df["emp_salary"]
df["increase_of_salary"] = (df["new_salary"] - df["previous_salary"]).round(2)

print("Salary Update Summary")
print(df[["role", "previous_salary", "new_salary", "increase_of_salary"]].to_string(index=False))
```
